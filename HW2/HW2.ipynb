```python
import pandas as pd

df=pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/Booli_sold.csv")
```

# Apartment Prices

### Task 1

Calculate ppsqm and add as a varible to the table.

```python
df['ppsq'] = round(df['soldPrice']/df['livingArea'])
df
```

### Task 2

Rank the top 5 most expensive appartments wrt ppsqm.

```python
df_top5 = df.sort_values(by='ppsq', ascending=False)
df_top5 = df_top5[:5]
df_top5
```

### Task 3

Average ppsqm in Ekhagen.

```python
ppsq = df['ppsq']

ppsq_avg = ppsq.mean()

ppsq_avg = f"{ppsq_avg:.0f}"
ppsq_avg
```

### Task 4 

I will add a new column with the % change from list price to sold price. It is a common varible to use in the context of the housing market. 

```python
df['% Change list vs sold'] = df['soldPrice']/df['listPrice']
df['% Change list vs sold'] = (df['% Change list vs sold']-1) * 100
df['% Change list vs sold'] = df['% Change list vs sold'].map("{:.2f}%".format)
df
```
