```python
ppsq = df['ppsq']

ppsq_avg = ppsq.mean()

ppsq_avg = f"{ppsq_avg:.0f}"
ppsq_avg
```

# Predicting Strokes

Task: Draw a conclusion using a suitable plot about stroke cases.

```python
import matplotlib.pyplot as plt
df_3=pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/stroke-data.csv")

data1 = df_3[['age', 'stroke']]

df_3['Age Group'] = pd.cut(df_3['age'], bins=range(0, 101, 10), right=False)

age_stroke_counts = df_3.groupby(['Age Group', 'stroke'], observed=False).size().unstack()

age_stroke_proportions = age_stroke_counts.div(age_stroke_counts.sum(axis=1), axis=0)

age_stroke_proportions.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title("Proportion of Stroke vs Non-Stroke by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Proportion of People")
plt.legend(["No Stroke", "Stroke"], title="Stroke Status")
plt.show()
```

From this dataset we can draw the conclusion that age and strokes are strongly correlated.
